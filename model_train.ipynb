```python
from fastai.vision.all import *
import matplotlib.pyplot as plt

# Set up paths to your dataset
train_path = '/Users/reetvikchatterjee/Desktop/SplitData/train'
valid_path = '/Users/reetvikchatterjee/Desktop/SplitData/valid'

path = Path(train_path)
dls = ImageDataLoaders.from_folder(path, valid_pct=0.2, item_tfms=Resize(128))


# Create a CNN learner (ResNet34 by default)
```

```python
learn = cnn_learner(dls, resnet34, metrics=[accuracy])

# Train the model (you can adjust epochs based on your need)
learn.fine_tune(4)

# Show confusion matrix
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()


# learn = load_learner('/path/to/save/model.pkl')

# Make predictions on new data (image file or video frame)
#img_path = '/path/to/new/image_or_video_frame.jpg'
#pred, pred_idx, probs = learn.predict(img_path)

#print(f"Prediction: {pred}, Confidence: {probs[pred_idx]:.4f}")
```
